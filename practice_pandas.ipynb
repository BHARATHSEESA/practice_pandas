```python
import pandas as pd
import numpy as np
```

```python
df  = pd.read_csv('transactions.csv')
df
```

## 1.What is the total sales amount across all transactions?

```python
#total amount spent by all customers
df['t_amt'].sum()
```

## 2.Which month had the highest total transaction amount?


```python
#month with the highest total amount spent
df["t_date"] = pd.to_datetime(df["t_date"])
df["Month"] = df["t_date"].dt.month
df.groupby("Month")["t_amt"].sum().sort_values(ascending=False).head(1)

```

## 3.What is the average transaction amount per customer?

```python
#average amount spent per customer
df.groupby('cust_id')['t_amt'].mean()
```

## 4.What is the trend of total sales over months?

```python
# Monthly sales trend
monthly_sales = df.groupby("Month")["t_amt"].sum()
monthly_sales
```

## 5.What is the highest single transaction amount recorded?


```python
#highest single transaction amount
df['t_amt'].max()
```

## 6.Which service category contributes the most to total revenue?


```python
#most popular service based on total amount spent
df.groupby('services')['t_amt'].sum().sort_values(ascending=False).head(1)
```

## 7.Which product generated the highest revenue?


```python
#most popular service based on total amount spent
df.groupby('products_used')['t_amt'].sum().sort_values(ascending=False).head(1)
```

## 8.How does the average transaction amount vary between services?


```python
#average amount spent per service
df.groupby("services")["t_amt"].mean()
```

## 9.How many unique customers made purchases?


```python
#number of unique customers
df['cust_id'].nunique()
```

## 10.Which customers have spent the most overall?


```python
#customer with the highest total amount spent
df.groupby("cust_id")["t_amt"].sum().sort_values(ascending=False).head(1)
```

## 11.What is the average number of transactions per customer?


```python
#average number of transactions per customer
df.groupby("cust_id").size().mean()
```

## 12.Are there customers who made multiple purchases in different categories?


```python
#number of unique services used by each customer
df.groupby("cust_id")["services"].nunique()
```

## 13.What percentage of customers are repeat buyers?


```python
#percentage of customers who made more than one transaction
repeat = df.groupby("cust_id").size()
(repeat[repeat > 1].count() / repeat.count()) * 100
```

## 14.Which product category has the highest total sales?


```python
#most popular product based on total amount spent
df.groupby('products_used')['t_amt'].sum().sort_values(ascending=False).head(1)
```

## 15.Which services are most popular by transaction count?

## 16.For each service, which product type is purchased most frequently?


```python
#most popular product-service combination based on total amount spent
df.groupby(["services","products_used"]).size().reset_index(name="count") \
  .sort_values(["services","count"], ascending=[True, False]) \
  .drop_duplicates("services")
```
